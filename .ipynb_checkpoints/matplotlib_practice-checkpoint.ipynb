```python
# Import dependencies
import matplotlib.pyplot as plt

# import matplotlib: imports the matplotlib dependency
# pyplot: a module that allows us to create charts
# plt: the alias
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a line chart using the MATLAB approach.
plt.plot(x_axis, y_axis)
```

```python
# Create a line chart using the Object-Oriented approach.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create a line chart using another Object-Oriented approach.
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create a line chart using yet another Object-Oriented approach.
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Create a line chart using the MATLAB approach and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create a line chart using the MATLAB approach and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create a line chart using the Object-Oriented approach, and add a label for the legend.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label = 'Boston')

# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")

# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)

# Create a title.
ax.set_title("PyBer Fare by Month")

# Add the legend.
ax.legend()

# Add the grid lines.
ax.grid()
```

```python
# Create a vertical bar chart using the MATLAB approach.
plt.bar(x_axis, y_axis)
```

```python
# Create a vertical bar chart using the MATLAB approach.
plt.bar(x_axis, y_axis, color="green", label="Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create the title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create a horizontal bar chart using the MATLAB approach.
plt.barh(x_axis, y_axis, color="magenta", label="Boston")

# Invert the y-axis data
plt.gca().invert_yaxis()

# Add a legend.
plt.legend()

# Add a title and axes labels.
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Assign 12 explode values for each wedge in the pie chart.
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Create a pie chart using the MATLAB method.
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')

# Increase the size of the pie chart.
plt.subplots(figsize=(8,8))

#plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Assign 12 explode values for each wedge in the pie chart.
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Create a pie chart using the Object-Oriented method.
fig, ax = plt.subplots()
ax.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%',
       shadow = "true",
       startangle = 265)

# Reverse the order so that January is at the top.
ax.invert_yaxis()

# Increase the size of the pie chart.
plt.subplots(figsize=(8,8))

plt.show()
```

```python
# Import statistics dependency needed to calculate standard deviation.
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standards deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Add error bars using the MATLAB approach.
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add error bars using the Object-Oriented approach.
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Add error bars to a bar chart using the MATLAB method.
plt.bar(x_axis, y_axis, yerr = stdev, capsize = 3)
```

```python
# Add error bars to a bar chart using the Object-Oriented method.
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr = stdev, capsize = 3)
```

```python
import numpy as np

# Change the major x-axis ticks on a horizontal bar chart using the MATLAB approach.
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# import numpy needed to use the arange function.
import numpy as np

# Create a range between 0 and 50 with an increment of 5.
np.arange(0, 51, step=5.0)

# Without "step=5.0", the range would return all the numbers from 0 to 50.
```

```python
# Change the major x-axis ticks on a horizontal bar chart using 
# the Object-Oriented approach.

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5))
```

```python
from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
```
